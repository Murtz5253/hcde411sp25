# Lab 2: Encodings and Marks in Altair (Solutions)

```python
import pandas as pd
import altair as alt
student_data = pd.DataFrame({'GPA': [4.0, 2.0, 0.4, 3.8, 3.0, 2.0, 1.8, 0.9, 3.3, 2.2, 4.0, 3.5, 2.4],
                             'SAT Score': [1600, 900, 800, 1600, 1400, 950, 900, 870, 1330, 400, 1550, 1450, 1000],
                             'Learning Enjoyment Rating': [5, 3, 2, 2, 4, 1, 3, 3, 4, 5, 3, 4, 1],
                             'Year': ['Freshman', 'Sophomore', 'Senior', 'Senior', 'Senior', 'Junior', 'Sophomore', 'Junior', 'Junior', 'Senior', 'Freshman', 'Sophomore', 'Junior'],
                             'Satisfaction': [5, 3, 1, 5, 4, 3, 2, 1, 4, 2, 5, 4, 2]
                            })

student_data
```

## Exercise 1: All these encodings

```python
alt.Chart(student_data).mark_point().encode(
    alt.X('GPA'),
    alt.Y('SAT Score'),
    alt.Size('Learning Enjoyment Rating'),
    alt.Color('Year'),
    alt.Shape('Year'),
    alt.Opacity('Satisfaction'),
    alt.Tooltip('GPA')
)
```

## Exercise 2: A couple more marks

```python
alt.Chart(student_data).mark_circle().encode(
    alt.X('GPA'),
    alt.Y('SAT Score')
)
```

```python
alt.Chart(student_data).mark_tick().encode(
    alt.X('GPA'),
    alt.Y('SAT Score')
)
```

## Exercise 3: Playing with color scales

```python
student_data2 = pd.DataFrame({'GPA': [4.0, 2.0, 0.4, 3.8, 3.0, 2.0, 1.8, 0.9, 3.3, 2.2, 4.0, 1.5, 2.4],
                             'SAT Score': [1600, 1600, 800, 500, 1400, 950, 1110, 870, 1330, 400, 550, 1450, 1000],
                             'Learning Enjoyment Rating': ['Very High', 'Medium', 'Low', 'Low', 'High', 'Very Low', 'Medium', 'Medium', 'High', 'Very High', 'Medium', 'High', 'Very Low']
                            })

alt.Chart(student_data2).mark_circle(size=200).encode(
    alt.X('GPA'),
    alt.Y('SAT Score'),
    alt.Color('Learning Enjoyment Rating:O', scale=alt.Scale(domain=['Very Low', 'Low', 'Medium', 'High', 'Very High'], scheme='greys'))
)
```
